```python
from project1_rag_chatbot.src.chunker import chunk_text
from project1_rag_chatbot.src.embedder import Embedder
from project1_rag_chatbot.src.retriever import Retriever

text = 'This is a short demo document about Apollo 11 and the Moon landing in 1969.'
chunks = chunk_text(text, chunk_size=10, overlap=2)
emb = Embedder()
vecs = emb.embed(chunks)
ret = Retriever()
ret.add(chunks, vecs)
qv = emb.embed(['Who walked on the Moon in 1969?'])[0]
print(ret.retrieve(qv, top_k=2))
```
